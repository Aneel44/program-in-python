```python
#Question.7 Write a python program which takes 5 input from user for different subject's marks,
#total it and generate marksheet using grades?
    
eng = int(input("Enter marks:"))
math = int(input("Enter marks:"))
python = int(input("Enter marks:"))
java = int(input("Enter marks:"))
history = int(input("Enter marks:"))
total = (eng+math+python+java+history)
grade = (total * 100)/500  

if grade >=80 and grade < 100:
    print(" Congrat! You obtained A+ or A-1 Grade.")
elif grade >=70 and grade < 80:
    print("You obtained A grade.")
elif grade >= 60 and grade < 70:
    print("You obtained B grade.")
elif grade >= 50 and grade < 60:
    print("You obtained C grade.")
elif grade >= 40 and grade < 50:
    print("You obtained D grade.")
else:
    print("Sorry! You are Fail try again!")
```
